**QUESTAO 3: Para a rede “USairport500”, calcule a entropia de Shannon da distribuição dos menores caminhos.
Use logaritmo na base 2 e considere apenas o maior componente da rede.**

a. 2,67

**b. 1,88**

c. 9,88

d. 0,12

e. 3,12

```python
from numpy  import *
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

G = nx.read_edgelist("/content/data/USairport500.txt")

N = len(G)
M = G.number_of_edges()
print('Number of nodes:', N)
print('Number of edges:', M)
```

```python
# Transformando em grafo sem direcao
G = G.to_undirected()

# Usando apenas o maior componente da rede
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])

# Convert the labels to integers starting with the index zero
G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
# Calcular a distribuição dos menores caminhos
vl = []
for i in np.arange(0,N):
    for j in np.arange(i+1, N):
        if(i != j):
            aux = nx.shortest_path(G, i, j)
            dij = len(aux)-1
            vl.append(dij)
```

```python
vl
```

```python
# Calcular a frequência de cada comprimento de caminho
path_length_counts = np.bincount(vl)

# Normalizar as contagens para obter a distribuição de probabilidade
path_length_probs = path_length_counts / path_length_counts.sum()

# Calcular a entropia de Shannon
shannon_entropy = -np.sum(path_length_probs[path_length_probs > 0] * np.log2(path_length_probs[path_length_probs > 0]))

print("Entropia de Shannon da distribuição dos menores caminhos:", "%3.2f"%shannon_entropy)
```
