**Questao 1: A média medida eigenvector centrality da rede USairport500 é igual a quanto? Considere apenas o maior componente.**

a. 0.64

b. 0.92

c. 0.45

d. 0.12

**e. 0.02**

```python
import numpy as np
import networkx as nx

G = nx.read_edgelist("/content/data/USairport500.txt")

N = len(G)
M = G.number_of_edges()
print('Number of nodes:', N)
print('Number of edges:', M)
```

```python
# A questao pediu para usar esses passos
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])
G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
EC = dict(nx.eigenvector_centrality(G, max_iter = 1000))
print('Eigenvetor centrality:', EC)
print('Eigenvetor centrality mean:', "%3.2f"%mean(list(EC.values())))
```

**Questao 2: Calcule a correlação de Pearson entre a medida betweeness centrality e grau para a rede hamsterster.**

**a. 0.8**

b. 0.5

c. 0.1

d. 0.4

e. 0.2

```python
import numpy as np
import networkx as nx

G = nx.read_edgelist("/content/data/hamsterster.txt")

N = len(G)
M = G.number_of_edges()
print('Number of nodes:', N)
print('Number of edges:', M)
```

```python
# A questao pediu para usar esses passos
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])
G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
# Betweeness centrality
B = dict(nx.betweenness_centrality(G))
B = list(B.values())
print('Betweeness centrality:', B)
```

```python
# Grau
vk = dict(G.degree())
vk = list(vk.values())
print('Degree centrality:', vk)
```

```python
# Correlação de Pearson
pearson = np.corrcoef(B, vk)
print('Pearson correlation:', "%3.1f"%pearson[0,1])
```

**Questao 3: Calcule o coeficiente de correlação de Spearman entre as medidas closeness centrality e k-core para a rede "Jazz".**

a. 1.0

b. 0.3

c. 0.5

d. 0.2

**e. 0.7**

```python
import numpy as np
import networkx as nx

G = nx.read_edgelist("/content/data/jazz.txt")

N = len(G)
M = G.number_of_edges()
print('Number of nodes:', N)
print('Number of edges:', M)
```

```python
# A questao pediu para usar esses passos
G = G.to_undirected()
G.remove_edges_from(nx.selfloop_edges(G))
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G = G.subgraph(Gcc[0])
G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
# Closeness centrality
CLC = dict(nx.closeness_centrality(G))
CLC = list(CLC.values())
print('Closeness centrality', CLC)
```

```python
# k-core
KC = dict(nx.core_number(G))
KC = list(KC.values())
print('k-core:', KC)
```

```python
# Correlação de Spearman
from scipy.stats import spearmanr
res, pval = spearmanr(CLC, KC)
print('Spearman correlation:', "%3.3f"%res)
```

*   Um vértice é considerado central se estiver conectado a outros centrais. É usada em redes dirigidas. -- **Pagerank**

*   É calculada a partir da exponencial da matriz de probabilidade de transição. -- **Random walk accessibility**

*   É uma medida de centralidade local, pois alguns vértices com valores altos dessa medida podem estar na borda da rede. -- **Degree centrality**

*   Considera o menores caminhos entre os pares de vértices. -- **Closeness centrality**

*   Usa os menores caminhos de modo a determinar a carga (em termos da quantidade de tráfego) em cada vértice. -- **Betweeness centrality**

*   É calculada a partir do espectro da matriz de adjacência. -- **Eigenvector centrality**
